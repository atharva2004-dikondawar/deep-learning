```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```python
img = image.load_img('train1/cat.9820.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img) # in opencv type is numpy array
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    # vertical_flip=True, # if true then the cat may lokk upside down which is false fact.
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='constant' # the image will have some blackness, if used reflect will reflect another
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i=0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):  #.flow is used for nparray of image stored in memory. 

    i=i+1

    if i == 10:
        break
    
```
