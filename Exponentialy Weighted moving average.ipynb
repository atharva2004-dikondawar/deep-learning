```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DF = pd.read_csv('DailyDelhiClimateTest.csv')
```

```python
DF=DF[['date','meantemp']]
DF.head()
```

```python
plt.scatter(DF['date'],DF['meantemp'],c='red')
plt.show()
```

```python
x1 = DF['meantemp'].ewm(alpha=0.9).mean()   # if alpha 0.9 then beta = 0.1.
```

```python
x1
```

```python
DF['ewma'] = x1
DF
```

```python
plt.scatter(DF['date'],DF['meantemp'],c='red')
plt.plot(DF['date'],x1,c='black',label='alpha = 0.9')
plt.legend()
plt.show()
```
